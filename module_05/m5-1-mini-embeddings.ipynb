```python
from openai import OpenAI
import numpy as np
from IPython.display import display, Markdown

def md(text):
    """Display text as rendered markdown."""
    display(Markdown(text))

client = OpenAI()
```

## 1. Creating Text Embeddings

Text embeddings convert text into numerical vectors that capture semantic meaning.

```python
# Sample texts
texts = [
    "The cat sits on the mat",
    "A feline rests on the carpet",
    "Python is a programming language",
    "The dog runs in the park"
]

# Create embeddings
def get_embedding(text):
    response = client.embeddings.create(
        model="text-embedding-3-small",
        input=text
    )
    return response.data[0].embedding

embeddings = [get_embedding(text) for text in texts]

md(f"**Created {len(embeddings)} embeddings**")
md(f"**Embedding dimensions:** {len(embeddings[0])}")
md(f"**First 5 values of embedding 1:** `{embeddings[0][:5]}`")
```

## 2. Cosine Similarity

Cosine similarity measures how similar two embeddings are (range: -1 to 1, where 1 is identical).

```python
def cosine_similarity(vec1, vec2):
    """Calculate cosine similarity between two vectors."""
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

# Compare all texts
output = "## Similarity Matrix\n\n"
output += "| | " + " | ".join([f"Text {i+1}" for i in range(len(texts))]) + " |\n"
output += "|" + "---|" * (len(texts) + 1) + "\n"

for i, text1 in enumerate(texts):
    row = f"| **Text {i+1}** |"
    for j, text2 in enumerate(texts):
        similarity = cosine_similarity(embeddings[i], embeddings[j])
        row += f" {similarity:.3f} |"
    output += row + "\n"

md(output)
```

```python
# Show which texts are most similar
md("### Original Texts:")
for i, text in enumerate(texts):
    md(f"{i+1}. {text}")
```

## 3. Finding Most Similar Text

Given a query, find the most similar text.

```python
query = "A cat lying on a rug"
query_embedding = get_embedding(query)

# Calculate similarities
similarities = [cosine_similarity(query_embedding, emb) for emb in embeddings]

# Find most similar
most_similar_idx = np.argmax(similarities)

md(f"**Query:** {query}")
md(f"\n**Most Similar Text:** {texts[most_similar_idx]}")
md(f"\n**Similarity Score:** {similarities[most_similar_idx]:.3f}")
md("\n### All Similarities:")
for i, (text, sim) in enumerate(zip(texts, similarities)):
    md(f"- {text}: **{sim:.3f}**")
```
