```python
# Install chromadb if needed
# !pip install chromadb openai

from openai import OpenAI
import chromadb
from IPython.display import display, Markdown

def md(text):
    """Display text as rendered markdown."""
    display(Markdown(text))

client = OpenAI()
```

## 1. Setup Vector Database

A vector database stores embeddings and enables fast similarity search.

```python
# Initialize ChromaDB (in-memory)
chroma_client = chromadb.Client()

# Create a collection
collection = chroma_client.create_collection(name="my_documents")

md("✅ Vector database initialized!")
```

## 2. Add Documents to Vector Database

Store documents with their embeddings.

```python
# Sample documents
documents = [
    "Python is a high-level programming language known for its simplicity.",
    "Machine learning is a subset of artificial intelligence.",
    "Vector databases enable efficient similarity search over embeddings.",
    "Natural language processing helps computers understand human language.",
    "Deep learning uses neural networks with multiple layers.",
    "RAG systems combine retrieval with language model generation."
]

def get_embedding(text):
    response = client.embeddings.create(
        model="text-embedding-3-small",
        input=text
    )
    return response.data[0].embedding

# Create embeddings and add to database
embeddings = [get_embedding(doc) for doc in documents]
ids = [f"doc_{i}" for i in range(len(documents))]

collection.add(
    embeddings=embeddings,
    documents=documents,
    ids=ids
)

md(f"✅ Added **{len(documents)} documents** to vector database")
```

## 3. Similarity Search

Query the vector database to find the most relevant documents.

```python
query = "How do I build AI applications that search knowledge bases?"

# Get query embedding
query_embedding = get_embedding(query)

# Search for similar documents
results = collection.query(
    query_embeddings=[query_embedding],
    n_results=3
)

md(f"**Query:** {query}")
md("\n### Top 3 Most Relevant Documents:")
for i, (doc, distance) in enumerate(zip(results['documents'][0], results['distances'][0])):
    md(f"\n**{i+1}.** {doc}")
    md(f"   - *Distance: {distance:.4f}*")
```

## 4. Try Different Queries

```python
queries = [
    "What is Python?",
    "Tell me about neural networks",
    "How do computers process text?"
]

for query in queries:
    query_embedding = get_embedding(query)
    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=1
    )
    
    md(f"**Query:** {query}")
    md(f"**Best Match:** {results['documents'][0][0]}")
    md("---")
```

## 5. Inspect Database Contents

```python
# Get all documents
all_docs = collection.get()

md(f"**Total documents in database:** {len(all_docs['documents'])}")
md("\n### All Documents:")
for i, doc in enumerate(all_docs['documents'], 1):
    md(f"{i}. {doc}")
```
