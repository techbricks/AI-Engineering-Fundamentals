````python
from IPython.display import display, Markdown

def md(text):
    """Display text as rendered markdown."""
    display(Markdown(text))
````

## 1. Why Chunking?

Documents are often too long to process as a single embedding. We need to split them into smaller chunks.

````python
# Sample long document
document = """Artificial intelligence (AI) has transformed many industries in recent years. 
Machine learning, a subset of AI, enables computers to learn from data without explicit programming. 
Deep learning uses neural networks with multiple layers to process complex patterns. 
Natural language processing (NLP) is a branch of AI that helps computers understand human language. 
Computer vision enables machines to interpret and understand visual information from the world. 
Reinforcement learning is a type of machine learning where agents learn by interacting with an environment.
Transfer learning allows models trained on one task to be adapted for another related task.
Large language models like GPT have revolutionized how we interact with AI systems."""

md(f"**Document length:** {len(document)} characters")
md(f"\n**Document:**\n\n{document}")
````

## 2. Basic Chunking by Character Count

Split document into fixed-size chunks.

````python
def chunk_by_characters(text, chunk_size):
    """Split text into chunks of fixed character size."""
    chunks = []
    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i + chunk_size])
    return chunks

chunk_size = 200
chunks = chunk_by_characters(document, chunk_size)

md(f"**Chunk size:** {chunk_size} characters")
md(f"**Number of chunks:** {len(chunks)}")
md("\n### Chunks:")
for i, chunk in enumerate(chunks, 1):
    md(f"\n**Chunk {i}:** (length: {len(chunk)})")
    md(f"```\n{chunk}\n```")
````

## 3. Chunking with Overlap

Add overlap between chunks to preserve context at boundaries.

````python
def chunk_with_overlap(text, chunk_size, overlap):
    """Split text into overlapping chunks."""
    chunks = []
    start = 0
    
    while start < len(text):
        end = start + chunk_size
        chunks.append(text[start:end])
        start = end - overlap  # Move back by overlap amount
        
        if end >= len(text):
            break
            
    return chunks

chunk_size = 200
overlap = 50
chunks_with_overlap = chunk_with_overlap(document, chunk_size, overlap)

md(f"**Chunk size:** {chunk_size} characters")
md(f"**Overlap:** {overlap} characters")
md(f"**Number of chunks:** {len(chunks_with_overlap)}")
md("\n### Chunks with Overlap:")
for i, chunk in enumerate(chunks_with_overlap, 1):
    md(f"\n**Chunk {i}:** (length: {len(chunk)})")
    md(f"```\n{chunk}\n```")
````

## 4. Visualizing Chunk Overlap

````python
# Show overlapping portions
if len(chunks_with_overlap) >= 2:
    chunk1 = chunks_with_overlap[0]
    chunk2 = chunks_with_overlap[1]
    
    # Find overlapping text
    overlap_text = chunk1[-overlap:]
    
    md("### Overlap Between Chunk 1 and Chunk 2:")
    md(f"\n**End of Chunk 1:**")
    md(f"```\n...{chunk1[-100:]}\n```")
    md(f"\n**Overlapping portion ({overlap} chars):**")
    md(f"```\n{overlap_text}\n```")
    md(f"\n**Start of Chunk 2:**")
    md(f"```\n{chunk2[:100]}...\n```")
````

------------------------------------------------------------------
💡 **Trade-off:** Larger chunks = more context but fewer chunks. Overlap = better context preservation but more storage.
-----------------------------------------------------------------



## 🎯 Summary

### Key Takeaways

1. **Text chunking** — splits long documents into smaller, manageable pieces for embedding
2. **Chunk overlap** — preserves context at boundaries between adjacent chunks
3. **Chunk sizing** — typical sizes range from 200–1000 characters; overlap is usually 10–20% of chunk size
4. **Retrieval quality** — chunking strategy directly affects how well a RAG system retrieves relevant information
